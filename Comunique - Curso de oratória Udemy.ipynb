# Comunique - Curso de oratória
##### Udemy

<br>

##### Exercício inicial: Linguagem corporal (Vídeo 1)
* Quem é você, qual é a sua história e de onde vem.
* O que você estuda e/ou trabalha?
* Por quê?
* Um fato inusitado ao seu respeito.

<br>

##### Como estabelecer uma conexão através do olhar: Transmitir mais credibilidade
* Olhe nos olhos.
* Quando estiver apresentando olhar para o grupo.
* Ao conhecer uma pessoa nova (Olhar nos olhos - cor deles, inclinar-se levemente para frente e aperta a mão com firmeza)

<br>

##### A posição da mulher maravilha
* Afastar as pernas, alinhando-as com os ombros.
* Endireitar a coluna.
* Levantar a cabeça.
* Expandir o peitoral.
* Ocupar o espaço ao nosso redor por meio da gesticulação. 
* Não fazer: Deixar as pernas juntas, permancer com a coluna encurvada, abaixa a cabeça e deixar os braços muito próximos ao corpo. 
* Exercício mulher maravilha (Vídeo 2)
* o que fazer com as mãos? Deixa-las a mostra
* Manter a posição da mulher maravilha com as mãos juntas em frente ao corpo.
* Gesticular a cabeça como se tivesse confirmando as informações dadas pelo outro orador.
* Alterar a visão ou contato visual entre o outro orador e a platéia.

<br>

##### Como evitar as distrações
* "Tudo tem que ter um motivo de ser" - Tem que ter uma motivação.
* Tem que movimentar na hora certa
* Mudando de assunto - Motivos para se movimentar.
* Ressaltar algo - Motivos para se movimentar.
* Evitar movimentos repetitivos e vícios de linguagem.
* Roupas que não chame atenção - Melhor usar cores neutras.

<br>

##### O poder do silêncio
* Não emitir sons para sinalizar que está pensando: EHH, AHH, UHH
* Tempo para pensarmos e tempo para os outros pensarem.
* Atiça a curiosidade das pessoas.
* Ressaltar o que estamos falando.

<br>

##### Gesticulação
* Pontos e vírculas da fala
* Sincronizada com a fala
* Movimentação: Braços, mão, pulsos e dedos
* Ajuda a ressalta um trecho da fala
* Abertura da gesticulação depende da distância da outra(s) pessoa(s)

<br>

##### Mímica
* Ajuda o entendimento do que está querendo passar e facilita a memorização do que foi falado.

<br>

##### Como treinar
* Feedback das pessoas
* Filmagem
* Anote os pontos que precisa melhorar quando treinar

<br>

##### Anotações e projeções
* Não pode roubar a atenção, fora de foco, deixar numa mão para baixo que não será usada na gesticulação.
* Será apenas um lembrete / consulta
* O foco tem que ser o que está falando e não o que está sendo projetado. 

<br>

##### O que fazer quando não prestarem atenção
* Levantar o braço
* Ficar em silêncio

<br>

##### Modulação de voz (Intensidade e tom - Agudo - Alto ou Grave - Baixo)
* Ressaltar, passar sentimento e passar um ideia
* Altere o tom entre agudo e grave
* Tom de voz grave + volume mais baixo = Conotação negativa
* Tom de voz agudo + volume mais alto = Conotação positiva

<br>

##### Expressão facial e corporal
* Levantar as sobrancelhas = Falar algo positivo
* Franzir as testa = Falar algo negativo
* Enverga para frente = Falar algo negativo

<br>

##### Técnicas de aquecimentos
* Visualização
* Conhecer o local com antecedência
* Respiração
* Fazer algo que acalme (Ouvir música, conversar ou ler um livro)
* Fingir que estamos confiantes (Fingir até que aconteça)

<br>

##### Dicas para sua voz
* Abrir mais a boca
* Respirar pela barriga
    * A - Policie
    * B - Colocar um peso na barriga e tente levanta-lo

### Técnicas / Elementos: <br>
* Atores
* Palhetas de cores
* Trilha sonora
* Continuidade
* Climax

<br>

##### Etapas:
* Começo: Introdução dos atores e cenários
* Meio: Aumento do elementos e climax (Exposição dos problemas e questionamentos)
* Final: Resolução dos problemas e questionamentos
##### OBS: Metáfora e analogias = Facilita a memorização da mensagem, ajuda a compreender idéias complexas, difícies e tem o "poder da aceitação".

<br>

##### Importância do convencimentos:
* Metáfora e analogias
* Exemplos
* Informações que passamos

<br>

##### Conhecer o público e tenha cuidado com a quantidade de informação (Passe as informações aos poucos e tenha um limite da quantidade) <br>
##### Organizar apresentação: Dividir em pequenas partes (Transitar entre as etapas de uma maneira fluida - Sem que as pessoas percebam) <br>

##### Etapa mais importante é o começo (Por que é a etapa que o público vai decidir se vão prestar atençãoou não):
* Aproveitar a expectativa do público
* Colocar imagens impactantes
* Colocar frases impactantes

<br>

##### Quando erramos:
* Erros pequenos: Fingir que nada aconteceu (Por que o público esquece) - Exemplos: Formatação, ortografia e palavras pequenas
* Erros grandes: Pedir desculpas, mencionar o erro e corrigir

<br>

##### Final:
* Colocar uma referência a algo do começo - Fechar o ciclo
* Plot twist (quebra de expectativas) - Induzir para o "A" e resultar em "B". Achar que vai acontecer "A" e acontece "B"
* Resumo / retrospectiva - Relembrar tudo o que foi aprendido

<br>

##### Criação de uma ligação:
* Usar humor na medida certa
* Sorriso
* Exemplos pessoais
* Expor uma "Fraqueza"
* Deve "suar" como uma "conversa" entre amigos
* Gerar sentimentos no público

<br>

##### Como preparar:
* Treinar / Testar
* Apresentar para outras pessoas (Diferentes)
* Gravar
* Feedback

### Storytelling <br>

* 1° mostrar os participantes (protagonistas e coadjuvantes), rotinas, rituais, ambiente e terminar com uma problemática/conflitos excitante, onde o protagonista tem que resolver.
* 2° resposta ao conflito inicial mais outros conflitos complexos.
* 3° Crise final existe um dilema que obriga a tomar uma decisão, climax onde a decisão coloca diante de confronto e resolução positivo ou negativa.

<br>

##### Metodologia: <br>
* Conflito: Enfrentamento entre duas forças opostas.
* Intriga: O que vai acontecer na próxima "cena"?
* Princípio impactante
* Detalhes
* Arquetipos: Vários tipos de protagonismo / Laços emocionais
* Metáforas
* Transformação do personagem
* 6 tipos de relatos: Transmissão de valores, relato visionário, relatos educativos, sei o que está pensando, quem sou e para que estou aqui.
* Final
* Artefatos visuais

<br>

##### Histórias têm o poder de nos mover, de nos sensibilizar e nos fazer acreditar numa ideia. Historia vendem e nós adoramos comprar.
* Escreva sobre o que fez, narrando seu processo e suas ideias.
* Comunicação de forma sutil, focando no propósito e no "por que": Empatia
* Documentar o conhecimento e transforma-los em narrativas
* Mostrando as dificuldades, expondo-se, permitindo ser vulnerável e se conectar de verdade por meio de uma voz comum e de objetivos parecidos.

<br>

##### Jornada do heroi:
1. O mundo comum: Detalhes capazes de fazer com que o público encontre pontos de identificação com ele. 
2. O chamado a aventura: Basta que seja um desafio que o obrigue a experimentar coisas novas.
3. Recusa do chamado: Compara a segunraça e conforto do seu lar com os caminhos tortuosos que poderá encontrar.
4. Encontro com o mentor: Empurrãozinho
5. A travessia do primeiro limiar: Algo desconhecido pelo personagem
6. Provas, aliados e inimigos: Deixando-o mais preparado para as maiores provações
7. Aproximação da caverna secreta: Retorna aos seus questionamentos iniciais e ao enfrentamento dos medos.
8. Aprovação: Significado de transformação
9. A recompensa
10. O caminho de volta: Escolha entre a realização de um objetivo pessoa ou um bem coletivo maior.
11. A ressurreição: Renascer para uma nova vida, totalmente transformada para todos.
12. O retorno com o elexir: Chegada ao seu local de origem simboliza o seu sucesso, conquista e mudança.
